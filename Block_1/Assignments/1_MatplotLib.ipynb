## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit

### The result should look something like this:
<img src="IMG/1.png" width=400>

```python
#plot benutzen, um den Graphen zu erstellen
plt.plot(data.month_number, data.total_profit)

#Titel hinzufuegen
plt.title('Company profit per month')

#y-Achsen
plt.yticks([100000, 200000, 300000, 400000, 500000])

#Achsenbeschriftungen
plt.xlabel('Month Number')
plt.ylabel('Total profit')

plt.show()
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
#Hilfsbariablen
liste_monat = data ['month_number']
facecream = data  ['facecream']
facewash = data ['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']

#Grafik erstellen
plt.plot(liste_monat, facecream, label = 'Face Cream', color = 'blue', marker = 'o')
plt.plot(liste_monat, facewash, label = 'Face Wash', color = 'y', marker = 'o')
plt.plot(liste_monat, toothpaste, label = 'Toothpaste', color = 'green', marker = 'o')
plt.plot(liste_monat, bathingsoap, label = 'Bathing Soap', color = 'red', marker = 'o')
plt.plot(liste_monat, shampoo, label = 'Shampoo', color = 'purple', marker = 'o')
plt.plot(liste_monat, moisturizer, label = 'Moisturizer', color = 'magenta', marker = 'o')

#Titel hinzufuegen
plt.title('Sales Data')

#Achsenbeschriftungen
plt.xlabel('Month Number')
plt.ylabel('Sold units in numberr')

#Legende lower right
plt.legend(loc = 2)

#y-Achsen und x-Achsen Skalierung
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xticks(liste_monat)

plt.show()
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.


### The result should look something like this:
<img src="IMG/4.png" width=400>

```python
liste_monat = data ['month_number']
toothpaste = data['toothpaste']

#Scatter Plot erstellen
plt.scatter(liste_monat, toothpaste, label= 'Tooth paste Sales data')

#Titel hinzufuegen
plt.title('Tooth paste sales data')

#Achsenbeschriftungen
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')

#Legende lower right
plt.legend(loc = 2)

#y-Achsen und x-Achsen Sklaierung
plt.yticks([4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000])
plt.xticks(liste_monat)

#grid-Layout
plt.grid(True, linestyle = '--')

plt.show()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
#Hilfsbariablen
liste_monat = data ['month_number']
facecream = data  ['facecream']
facewash = data ['facewash']

#Bar Chart erstellen
plt.bar([w-0.25 for w in liste_monat], facecream, label= 'Face Cream sales data', width = 0.25, align = 'edge' )
plt.bar([w+0.25 for w in liste_monat], facewash, label = 'Face Wash sales data', width = -0.25, align = 'edge' )

#Achsenbeschriftungen
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

#Titel hinzufuegen
plt.title('Facewash and facecream sales data')

#Legende lower right
plt.legend(loc = 2)

#y-Achsen und x-Achsen Sklaierung
plt.yticks([500, 1000, 1500, 2000, 2500, 3000, 3500])
plt.xticks(liste_monat)

#grid-Layout
plt.grid(True, linestyle = '--')

plt.show()
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 


### The result should look something like this:
<img src="IMG/6.png" width=400>

```python
data['toothpaste'].sum()
```

```python
labels = 'Toothpaste',  'Bathing Soap', 'Shampoo', 'Moisturizer', 'FaceCream', 'FaceWash',
sizes = [data['toothpaste'].sum(), data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum(), 
         data['facecream'].sum(), data['facewash'].sum(),]
plt.axis('equal')
plt.pie(sizes, labels = labels, autopct='%1.1f%%')

plt.legend(loc='lower right')

plt.title('Sales data')
plt.show()
```
